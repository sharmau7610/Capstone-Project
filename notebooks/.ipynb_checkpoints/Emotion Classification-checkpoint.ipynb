## <p style="background-color:yellow; font-family:newtimeroman; font-size:150%; text-align:center">Introduction</p>

Emotion Analysis aims to detect and recognize types of feelings through the expression of texts, such as anger, disgust, fear, happiness, sadness, and surprise.

![emotions](https://s01.sgp1.cdn.digitaloceanspaces.com/article/107360-byxhzxvmlb-1544009839.jpg)


## <p style="background-color:yellow; font-family:newtimeroman; font-size:150%; text-align:center">About Dataset</p>

This dataset contains emotions expressed through text and the emotions included in the data are ["joy","sadness","anger","fear","disgust","shame","neutral","surprise"].

The csv file consist of text descriptions and its corresponding label.

## <p style="background-color:yellow; font-family:newtimeroman; font-size:150%; text-align:center">Problem Statement</p> 

Since it deals with the text and we are classifying the text into respective emotions(labels) this is clearly a NLP text classfication problem. We have multiple emotions(labels) in our dataset hence it is a multi-classification problem.

```python
## Sentiment Analysis
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment>0:
        result='positive'
    elif sentiment<0:
        result='negative'
    else:
        result='neutral'
    return result
```

```python
df['Sentiment']=df['Text'].apply(get_sentiment)
```

```python
def cont_to_exp(x):
    if type(x) is str:
        for key in contractions:
            value = contractions[key]
            x = x.replace(key,value)
        return x
    else:
        return x
```

```python
# User handles
df['Clean_Text'] = df['Text'].apply(nfx.remove_userhandles)
#lower
df['Clean_Text'] = df['Clean_Text'].apply(lambda x: x.lower())
# contractions
df['Clean_Text'] = df['Clean_Text'].apply(lambda x: cont_to_exp(x))
# Stopwords
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_stopwords)
# punctuations'
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_punctuations)
# HTML TAGS
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_html_tags)
# urls
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_urls)
# Emails
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_emails)
```

```python
df = df[~(df['Clean_Text'] == ' ')]
df = df[~(df['Clean_Text'] == '')]
print(df.shape)
df.head()
```

```python
## KeyWord Extraction and Word Cloud of top 4 Emotions
```

```python
from wordcloud import WordCloud
```

```python
from collections import Counter
```

```python
def extract_keywords(text,num=25):
    tokens = [t for t in text.split()]
    most_common_tokens = Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
emotion_list = df['Emotion'].unique().tolist()
emotion_list
```

```python
def common_words_vis(mydict,emotion_name):
    df_1 = pd.DataFrame(mydict.items(),columns=['token','count'])
    plt.figure(figsize=(18,9))
    plt.title("Plot of {} top 25 most common KeyWords".format(emotion_name))
    sns.barplot(x='token',y='count',data=df_1)
    plt.xticks(rotation=45)
    plt.show();
```

## <p style="background-color:yellow; font-family:newtimeroman; font-size:150%; text-align:left">3.Model Building</p>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
import lightgbm as lgb

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.preprocessing import LabelEncoder

import optuna
```

```python
##Label Encoding
def label_encode(data,label):
    labels=data[label].map(
    {
        "joy":0,
        "sadness":1,
        "fear":2,
        "anger":3,
        "surprise":4,
        "shame":5,
        "neutral":6,
        "disgust":7,
    }
    )
    return labels
```

```python
df["Emotion"]=label_encode(df,"Emotion")
df.head()
```

```python
df = df[['Emotion','Clean_Text']]
df.head()
```

```python
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

```python
Xfeatures
```

```python
X_train,X_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.20,random_state=30)
```

```python
## Type1 - Count Vectorizer
```

```python
cv = CountVectorizer(max_features=20000)
X_vec = cv.fit_transform(X_train)
X_test_vec = cv.transform(X_test)
```

```python
## logistic reg 1
model_lr=LogisticRegression()
model_lr.fit(X_vec,y_train)
y_preds = model_lr.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Naive bayes
model_nb=MultinomialNB()
model_nb.fit(X_vec,y_train)
y_preds = model_nb.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Random Forest
model_rf=RandomForestClassifier()
model_rf.fit(X_vec,y_train)
y_preds = model_rf.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Xgboost
model_xgb=XGBClassifier(n_estimators = 1000,random_state=10)
model_xgb.fit(X_vec,y_train)
y_preds = model_xgb.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Logistic Reg 2
model_lr2=LogisticRegression(random_state=10,max_iter=1500)
model_lr2.fit(X_vec,y_train)
y_preds = model_lr2.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Logistic Reg 3
model_lr3=LogisticRegression(random_state=10,max_iter=1500,solver='liblinear',penalty='l2')
model_lr3.fit(X_vec,y_train)
y_preds = model_lr3.predict(X_test_vec)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
# Save Model-lr+cv
import joblib
logisticreg2 = open("emotion_clf_lr2cv.pkl","wb")
joblib.dump(model_lr2,logisticreg2)
logisticreg2.close()
```

```python
# Save Model-lr+cv
import joblib
logisticreg3 = open("emotion_clf_lr3cv.pkl","wb")
joblib.dump(model_lr3,logisticreg3)
logisticreg3.close()
```

```python
## Type 2 -TFIDF
```

```python
vect_tf = TfidfVectorizer()
X_vec_tf = vect_tf.fit_transform(X_train)
X_test_vec_tf = vect_tf.transform(X_test)
```

```python
## Random Forest
model_tf_rf=RandomForestClassifier(random_state=10)
model_tf_rf.fit(X_vec_tf,y_train)
y_preds = model_tf_rf.predict(X_test_vec_tf)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```

```python
## Xgboost
model_tf_xgb=XGBClassifier(random_state=10)
model_tf_xgb.fit(X_vec_tf,y_train)
y_preds = model_tf_xgb.predict(X_test_vec_tf)
acc_score = accuracy_score(y_test,y_preds)
print(acc_score)
```
